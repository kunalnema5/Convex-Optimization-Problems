# **Part (a)**

To minimize the initial investment we have to minimize both the funding of expense stream through bank account and investment types.

*   Total initial investment = b1 + Σxⱼ
*   Objective = minimize(b1 + Σxⱼ)

We have to minimize the objective under the constraints:


*   b≥0 & w≥0 as bank balance and money withdrawals cannot be negative
*   xⱼ≥0 ∀ j=1,..,m (given)
*   wₜ + (Px)ₜ ≥ eₜ as total funding after each time period (through bank account and investment payouts) must cover our expenditure at time t.
*   b[t+1] = (1 + ρ)bₜ − wₜ as we have to update bank balance after each time peroid

Thus minimizing objective under given constraints is a convex optimization problem.

# **Part (b)**

```python
# Part (b)
import cvxpy as cp

x = cp.Variable(m)
b = cp.Variable(n)
w = cp.Variable(n)

constraints = []
for t in range(0, n-1):
  constraints += [b[t+1] == (1+rho)*b[t] - w[t]]

constraints += [x>=0, b>=0, w>=0, (cp.reshape(w, (n,1)) + cp.reshape((P @ x), (n,1))) >= e]

objective = cp.Minimize(b[0] + np.ones(m).T @ x)

# Minimum initial investment
prob = cp.Problem(objective, constraints)
min_invest = prob.solve()
print("Minimum initial investment: ", min_invest)

# When no investments are purchased then we have an extra constraints as x=0 
constraints += [x==0]
prob = cp.Problem(objective, constraints)
investment = prob.solve()
print("Initial investment when no investments are purchased :", investment)
```
